# 가상화폐 시세 판단하기

> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.

|Key Name        |     Description|
|---|---|
|opeing_price    |최근 24시간 내 시작 거래금액|
|closing_price   |최근 24시간 내 마지막 거래금액|
|min_price       |최근 24시간 내 최저 거래금액|
|max_price       |최근 24시간 내 최고 거래금액|

````python
import requests
btc = requests.get('https://api.bithumb.com/public/ticker/btc').json()['data']
print(btc)
````

````python
# 아래에 코드를 작성하세요.
var=int(btc['max_price'])-int(btc['min_price'])
o_var=int(btc['opening_price'])+var
if o_var>int(btc['max_price']):
    print("상승장")
else:
    print("하락장")
````

````python
#교수님 답
opening_price=int(btc['opening_price'])
max_price=int(btc['max_price'])
min_price=int(btc['min_price'])

if opening_price +(max_price-min_price)>max_price:
    print('상승장')
else:
    print('하락장')
````

````python
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']
print(currency)
````

````python
# 아래에 코드를 작성하세요.
for i in currency:
    stock=currency[i]
    if stock==[]:
        print(f'{i} 값 없음')
        print('\n')
        continue
    elif i=='date':
        print(f'날짜 {currency[i]}')
    else:
        var=float(stock['max_price'])-float(stock['min_price'])
        o_var=float(stock['opening_price'])+var
        if o_var>float(stock['max_price']):
            print(f"{i} 상승장")
        else:
            print(f"{i} 하락장")
        print('\n')
````
